```python
# Calculate fragment logP

from rdkit import Chem
from rdkit.Chem import Descriptors

import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
from scipy.stats import ttest_rel
import matplotlib as plt
FILE="KI_Gygi 2021 HCT116 CL_AC fragment.xlsx"


df = pd.read_excel(FILE,index_col=0)

#remove proteins from contaminant database
#contaminants=["ALBU", "AMY1A", "AMY1B", "AMY1C", "ANT3", "ANXA5", "B2MG", "BID", "CAH1", "CAH2", "CATA", "CATD", "CATG", "CO5", "CRP", "CYB5", "CYC", "EGF", "FABPH", "GELS", "GSTA1", "GSTP1", "HBA", "HBB", "HBEGF", "IGF2", "IL8", "INHBA", "INHBB", "KCRM", "LALBA", "LEP", "LYSC", "MYG", "NEDD8", "NQO1", "NQO2", "PDGFB", "PPIA", "PRDX1", "RASH", "RET4", "RS27A", "SODC", "SUMO1", "HARS1", "TAU", "THIO", "TNFA", "TRFE", "TRFL", "UB2E1", "UBE2C", "K1C26", "K2C3", "K1C15", "K2C79", "K2C6B", "K1C17", "K2C4", "K2C73", "K2C71", "K2C7", "K2C8", "K1C39", "K1C18", "K1C28", "K1C16", "K2C1", "K2C5", "K2C80", "K2C1B", "K2C75", "K2C6A", "K2C72", "K1C24", "K1C19", "K2C74", "K1C27", "K1C20", "K1C9", "K1C23", "K1C12", "K1C14", "K2C6C", "K1C10", "K1C13", "K22O", "K1C25", "K2C78", "K22E", "K1C40", "KRT85", "KRT38", "KRT34", "KRT86", "KRT35", "KT33B", "KRT81", "KRT37", "KT33A", "KRT83", "K1H1", "KRT82", "K1H2", "KRT36", "KRT84"]
#df = df[~df.index.isin(contaminants)]
df.to_csv("test.csv")
df = -np.log10(df)


cmpd_df= pd.read_excel("41587_2020_778_MOESM7_ESM_compoundsInfo.xlsx", index_col=0)
logP_list=[]
for smile in cmpd_df["SMILES"]:
    mol=Chem.MolFromSmiles(smile)
    logP=Descriptors.MolLogP(mol)
    logP_list.append(logP)
cmpd_df["logP"]=logP_list

cmpd_df
```

```python
# Calculate dG = RTlnK
R=8.3145 #J/(K*mol)
T=310.15 #K at 37deg
df_dG = R*T*np.log(10**-df)/1000/4.184 #kcal/mol

df_LE = pd.DataFrame(index=df_dG.index)

for fragment in df_dG.columns:
    atom_number=cmpd_df["Atom number"][fragment]
    df_LE[fragment]=-df_dG[fragment]/atom_number

df_LE.to_csv("LE_CL_AC.csv")
df_LE.head
```

```python
import re

# Function to extract numerical value from the CL Name
def natural_key(text):
    return [int(s) if s.isdigit() else s for s in re.split(r'(\d+)', text)]

# Pivot the DataFrame
pivot_df_CL = new_df_CL.pivot(index='Index', columns='CL Name', values='Difference')

# Sort the columns in natural order
ordered_columns = sorted(pivot_df_CL.columns, key=natural_key)
pivot_df_CL = pivot_df_CL[ordered_columns]

# Reset the column index and remove the name from columns (if needed)
pivot_df_CL = pivot_df_CL.reset_index().rename_axis(None, axis=1)

print(pivot_df_CL)

pivot_df_CL.to_csv("LLE_CL_2D.csv")
```
